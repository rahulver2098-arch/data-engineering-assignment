```python
import pandas as pd
import os

# ---------------------------
# Configuration
# ---------------------------
RAW_DATA_PATH = r"C:\Users\User\OneDrive\Documents\data engineering assignment\data"
PROCESSED_DATA_PATH = r"C:\Users\User\OneDrive\Documents\data engineering assignment\processed"

USERS_FILE = "users_100.csv"
ORDERS_FILE = "orders_500.csv"

# ---------------------------
# Ensure directories exist
# ---------------------------
os.makedirs(RAW_DATA_PATH, exist_ok=True)
os.makedirs(PROCESSED_DATA_PATH, exist_ok=True)

# ---------------------------
# Ingestion Functions
# ---------------------------
def ingest_users():
    print("Ingesting users data...")
    users_df = pd.read_csv(os.path.join(RAW_DATA_PATH, USERS_FILE))
    users_df.to_csv(
        os.path.join(PROCESSED_DATA_PATH, "users_staging.csv"),
        index=False
    )
    print(f"Users records ingested: {len(users_df)}")


def ingest_orders():
    print("Ingesting orders data...")
    orders_df = pd.read_csv(os.path.join(RAW_DATA_PATH, ORDERS_FILE))
    orders_df.to_csv(
        os.path.join(PROCESSED_DATA_PATH, "orders_staging.csv"),
        index=False
    )
    print(f"Orders records ingested: {len(orders_df)}")


# ---------------------------
# Main Execution
# ---------------------------
if __name__ == "__main__":
    print("Starting data ingestion process...\n")

    ingest_users()
    ingest_orders()

    print("\nData ingestion completed successfully.")
```
